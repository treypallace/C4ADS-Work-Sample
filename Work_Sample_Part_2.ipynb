**Data Sources** <br>
All data ranges are from 1/1/2015 to 12/31/2020  to encompass a full five years of data (with the exception of RACE, which went public in 10/2015). Data was recorded on a monthly basis on the first day of each month.

**Commodities** <br>
All commodity data was selected as a Producer Price Index by Commodity
* [Steel](https://fred.stlouisfed.org/series/WPU101707)
* [Rubber Tires](https://fred.stlouisfed.org/series/WPU071201)
* [Glass](https://fred.stlouisfed.org/series/WPU13110511)
* [Aluminum](https://fred.stlouisfed.org/series/PCU3313153313150)

**Stocks**
* [Tesla (TSLA)](https://finance.yahoo.com/quote/TSLA/history?period1=1420070400&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)
* [General Motors (GM)](https://finance.yahoo.com/quote/GM/history?period1=1420070400&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)
* [Ford (F)](https://finance.yahoo.com/quote/F/history?period1=1420070400&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)
* [Honda (HMC)](https://finance.yahoo.com/quote/HMC/history?period1=1420070400&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)
* [Ferrari (RACE)](https://finance.yahoo.com/quote/RACE/history?period1=1447200000&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)

**Indexes**
* [Global Auto Index (CARZ)](https://finance.yahoo.com/quote/CARZ/history?period1=1420070400&period2=1609372800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true)



```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

# Importing Data

Import data on commodity prices

```python
steel = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/SteelCommodityData.csv")
rubber = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/TiresCommodityData.csv")
glass = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/GlassCommodityData.csv")
aluminum = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/AluminumCommodityData.csv")
```

Import data on stock prices

```python
tesla = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/TSLA.csv")
gm = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/GM.csv")
ford = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/F.csv")
carz = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/CARZ.csv")
honda = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/HMC.csv")
ferrari = pd.read_csv("https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/RACE.csv")
```

Read in 2021 data - this is what we will compare 2021 predictions to

```python
steel2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/Steel2021Data.csv')
steel2021 = pd.concat([steel, steel2021])

rubber2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/Rubber2021Data.csv')
rubber2021 = pd.concat([rubber, rubber2021])

glass2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/Glass2021Data.csv')
glass2021 = pd.concat([glass, glass2021])

aluminum2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/Aluminum2021Data.csv')
aluminum2021 = pd.concat([aluminum, aluminum2021])

tesla2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/TSLA2021.csv')
tesla2021 = pd.concat([tesla, tesla2021])

ford2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/F2021.csv')
ford2021 = pd.concat([ford, ford2021])

gm2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/GM2021.csv')
gm2021 = pd.concat([gm, gm2021])

honda2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/HMC2021.csv')
honda2021 = pd.concat([honda, honda2021])

ferrari2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/RACE2021.csv')
ferrari2021 = pd.concat([ferrari, ferrari2021])

carz2021 = pd.read_csv('https://raw.githubusercontent.com/cawarburton5/ForgeNode/main/CARZ2021.csv')
carz2021 = pd.concat([carz, carz2021])
```

# Data Cleaning

Create indicator columns

```python
# Stocks
tesla['id'] = 'Tesla (TSLA)'
gm['id'] = 'GM'
ford['id'] = 'Ford (F)'
honda['id'] = 'Honda (HMC)'
ferrari['id'] = 'Ferrari (RACE)'

tesla2021['id'] = 'Tesla (TSLA)'
gm2021['id'] = 'GM'
ford2021['id'] = 'Ford (F)'
honda2021['id'] = 'Honda (HMC)'
ferrari2021['id'] = 'Ferrari (RACE)'

# Index
carz['id'] = 'Global Auto Index (CARZ)'

carz2021['id'] = 'Global Auto Index (CARZ)'

# Commodities
steel['id'] = 'Steel'
rubber['id'] = 'Rubber'
glass['id'] = 'Glass'
aluminum['id'] = 'Aluminum'

steel2021['id'] = 'Steel'
rubber2021['id'] = 'Rubber'
glass2021['id'] = 'Glass'
aluminum2021['id'] = 'Aluminum'
```

```python
# Stocks
tesla['type'] = 'Stock'
gm['type'] = 'Stock'
ford['type'] = 'Stock'
honda['type'] = 'Stock'
ferrari['type'] = 'Stock'

tesla2021['type'] = 'Stock'
gm2021['type'] = 'Stock'
ford2021['type'] = 'Stock'
honda2021['type'] = 'Stock'
ferrari2021['type'] = 'Stock'

# Index
carz['type'] = 'Index'

carz2021['type'] = 'Index'

# Commodities
steel['type'] = 'Commodity'
rubber['type'] = 'Commodity'
glass['type'] = 'Commodity'
aluminum['type'] = 'Commodity'

steel2021['type'] = 'Commodity'
rubber2021['type'] = 'Commodity'
glass2021['type'] = 'Commodity'
aluminum2021['type'] = 'Commodity'
```

Rename columns

```python
steel.rename(columns={'DATE':'date', 'WPU101707':'price'}, inplace=True)
rubber.rename(columns={'DATE':'date', 'WPU071201':'price'}, inplace=True)
glass.rename(columns={'DATE':'date', 'WPU13110511':'price'}, inplace=True)
aluminum.rename(columns={'DATE':'date', 'PCU3313153313150':'price'}, inplace=True)

steel2021.rename(columns={'DATE':'date', 'WPU101707':'price'}, inplace=True)
glass2021.rename(columns={'DATE':'date', 'WPU13110511':'price'}, inplace=True)
rubber2021.rename(columns={'DATE':'date', 'WPU071201':'price'}, inplace=True)
aluminum2021.rename(columns={'DATE':'date', 'PCU3313153313150':'price'}, inplace=True)
```

# Joining Data

Concatenate stock and commodities dataframes

```python
# Combine stock dataframes
stocks = pd.concat([tesla, gm, ford, carz, honda, ferrari])
stocks2021 = pd.concat([tesla2021, gm2021, ford2021, carz2021, honda2021, ferrari2021])

# Combine commodities dataframes
commodities = pd.concat([steel, rubber, glass, aluminum])
commodities2021 = pd.concat([steel2021, rubber2021, glass2021, aluminum2021])

# Create a stocks dataframe that matches commodities
stocks_for_concat = stocks[['Date', 'Close', 'id','type']]
stocks_for_concat.rename(columns={'Date':'date', 'Close':'price', 'ID':'id','Type':'type'}, inplace=True)

stocks_for_concat2021 = stocks2021[['Date', 'Close', 'id','type']]
stocks_for_concat2021.rename(columns={'Date':'date', 'Close':'price', 'ID':'id','Type':'type'}, inplace=True)

# Concatenate stock and commodities data
prices = pd.concat([stocks_for_concat, commodities])
prices2021 = pd.concat([stocks_for_concat2021, commodities2021])

# Adjust date to a datetime format
prices['date'] = pd.to_datetime(prices['date'])
prices2021['date'] = pd.to_datetime(prices2021['date'])
prices2021['dataset'] = 'Actual'
```

```python
p = Prophet()

# Steel prediction
steel_df = prices[prices['id'] == 'Steel']

# Prophet requires that the dataframe has columns "ds" for dates and "y" for the response
steel_df.rename(columns={'date':'ds', 'price':'y'}, inplace=True)
p_steel = p.fit(steel_df)

future_dates = p_steel.make_future_dataframe(periods=10, freq='M', include_history=True)
# ds=date, yhat=predicted values, yhat_lower/upper=uncertainty intervals
steel_forecast = p_steel.predict(future_dates)

# Combine actual values with predicted values
steel_forecast['dataset'] = 'Predicted'
steel_forecast['id'] = 'Steel'
steel_forecast['type'] = 'Commodity'
steel_forecast.rename(columns={'ds':'date', 'yhat':'price'}, inplace=True)

# Plot Actual vs. Predicted
steel_comparison = pd.concat([prices2021[prices2021['id'] == 'Steel'][['date', 'price', 'dataset']], steel_forecast[['date', 'price', 'dataset']]])

steel_fig = px.line(steel_comparison, x='date', y='price', color='dataset',
                    labels = {'pctchange':'Percent Change','date':'Date','id':'ID','type':'Type'},
                    title = '<b>Projected Steel Prices vs. Actual Steel Prices</b>',
                    width=900, height=450, error_y=None
                    )
steel_fig
```
